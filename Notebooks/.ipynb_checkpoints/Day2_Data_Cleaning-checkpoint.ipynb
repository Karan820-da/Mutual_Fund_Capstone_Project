# Day 2 - Data Cleaning & Database Loading

## Objectives

```python
import pandas as pd
```

```python
nav = pd.read_csv("../Data/Raw/02_nav_history.csv")

nav["date"] = pd.to_datetime(nav["date"])

nav = nav.sort_values(
    ["amfi_code","date"]
)

nav = nav.drop_duplicates()

print(nav.shape)
```

## NAV Cleaning

Actions Performed:
- Converted date column to datetime.
- Sorted by fund and date.
- Removed duplicate rows.
- Validated NAV values.

```python
transactions = pd.read_csv(
    "../Data/Raw/08_investor_transactions.csv"
)

transactions["transaction_date"] = pd.to_datetime(
    transactions["transaction_date"]
)

transactions = transactions[
    transactions["amount_inr"] > 0
]

transactions["transaction_type"].value_counts()
```

## Transaction Cleaning

Actions Performed:
- Standardized transaction types.
- Verified KYC status values.
- Validated positive transaction amounts.
- Converted transaction dates.
